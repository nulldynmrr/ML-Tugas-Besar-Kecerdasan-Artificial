```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.metrics import accuracy_score
```

```python
x = pd.read_csv("data.csv")
```

```python
# Rentang variabel
x_income = np.arange(0, 200001, 1000)      # Income: 0 - 200000
x_age = np.arange(18, 71, 1)               # Age: 18 - 70
x_credit = np.arange(0, 101, 1)            # Credit Score (0-100 scale)
x_eligibility = np.arange(0, 101, 1)       # Output: Loan eligibility (0-100 scale)
```

```python
def income_low(x_income):
    if x_income <= 0:
        return 1
    elif x_income <= 70000:
        return (70000 - x_income)/(70000 - 0)
    else:
        return 0

def income_med(x_income):
    if x_income <= 50000:
        return 0
    elif x_income <= 100000:
        return (x_income - 50000)/(100000 - 50000)
    elif x_income <= 150000:
        return (150000 - x_income)/(150000 - 100000)
    else:
        return 0

def income_high(x_income):
    if x_income <= 100000:
        return 0
    elif x_income <= 200000:
        return (x_income - 100000)/(200000 - 100000)
    else:
        return 1
```

```python
def age_young(x_age):
    if x_age <= 18:
        return 1
    elif x_age <= 30:
        return (30 - x_age)/(30 - 18)
    else :
        return 0

def age_middle(x_age):
    if x_age <= 25:
        return 0
    elif x_age <= 40:
        return (x_age - 25)/(40 - 25)
    elif x_age <= 55:
        return (55 - x_age)/(55 - 40)
    else :
        return 0

def age_old(x_age):
    if x_age <= 50:
        return 0
    elif x_age <= 70:
        return (x_age - 50)/(70 - 50)
    else :
        return 1
```

```python
def credit_low(x_credit):
    if x_credit <= 0:
        return 1
    elif x_credit <= 40:
        return (40 - x_credit)/(40 - 0)
    else :
        return 0

def credit_med(x_credit):
    if x_credit <= 30:
        return 0
    elif x_credit <= 50:
        return (x_credit - 30)/(50 - 30)
    elif x_credit <= 70:
        return (70 - x_credit)/(70 - 50)
    else :
        return 0

def credit_high(x_credit):
    if x_credit <= 60:
        return 0
    elif x_credit <= 100:
        return (x_credit - 60)/(100 - 60)
    else :
        return 1
```

```python
def eligibility_no(x_eligibility):
    if x_eligibility <= 0:
        return 1
    elif x_eligibility <= 40:
        return (40 - x_eligibility)/(40 - 0)
    else :
        return 0

def eligibility_maybe(x_eligibility):
    if x_eligibility <= 30:
        return 0
    elif x_eligibility <= 50:
        return (x_eligibility - 30)/(50 - 30)
    elif x_eligibility <= 70:
        return (70 - x_eligibility)/(70 - 50)
    else :
        return 0

def eligibility_yes(x_eligibility):
    if x_eligibility <= 60:
        return 0
    elif x_eligibility <= 100:
        return (x_eligibility - 60)/(100 - 60)
    else :
        return 1
```

```python
# --- Hitung Derajat Keanggotaan untuk Seluruh Rentang Variabel ---

# Income
y_income_low = [income_low(val) for val in x_income]
y_income_med = [income_med(val) for val in x_income]
y_income_high = [income_high(val) for val in x_income]

# Age
y_age_young = [age_young(val) for val in x_age]
y_age_middle = [age_middle(val) for val in x_age]
y_age_old = [age_old(val) for val in x_age]

# Credit Score
y_credit_low = [credit_low(val) for val in x_credit]
y_credit_med = [credit_med(val) for val in x_credit]
y_credit_high = [credit_high(val) for val in x_credit]

# Eligibility
y_eligibility_no = [eligibility_no(val) for val in x_eligibility]
y_eligibility_maybe = [eligibility_maybe(val) for val in x_eligibility]
y_eligibility_yes = [eligibility_yes(val) for val in x_eligibility]


# --- Visualisasi Menggunakan Matplotlib dengan Subplot ---
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(14, 10))

# Plot Income
axes[0, 0].plot(x_income, y_income_low, 'b', label='Low')
axes[0, 0].plot(x_income, y_income_med, 'g', label='Medium')
axes[0, 0].plot(x_income, y_income_high, 'r', label='High')
axes[0, 0].set_title('Income')
axes[0, 0].set_xlabel('Income (IDR)')
axes[0, 0].set_ylabel('Degree of Membership')
axes[0, 0].legend()
axes[0, 0].grid(True)
axes[0, 0].set_ylim(-0.1, 1.1)
axes[0, 0].set_xticks(np.arange(0, 200001, 50000))
axes[0, 0].ticklabel_format(style='plain', axis='x')


# Plot Age
axes[0, 1].plot(x_age, y_age_young, 'b', label='Young')
axes[0, 1].plot(x_age, y_age_middle, 'g', label='Middle')
axes[0, 1].plot(x_age, y_age_old, 'r', label='Old')
axes[0, 1].set_title('Age')
axes[0, 1].set_xlabel('Age (Years)')
axes[0, 1].set_ylabel('Degree of Membership')
axes[0, 1].legend()
axes[0, 1].grid(True)
axes[0, 1].set_ylim(-0.1, 1.1)


# Plot Credit Score
axes[1, 0].plot(x_credit, y_credit_low, 'b', label='Low')
axes[1, 0].plot(x_credit, y_credit_med, 'g', label='Medium')
axes[1, 0].plot(x_credit, y_credit_high, 'r', label='High')
axes[1, 0].set_title('Credit Score')
axes[1, 0].set_xlabel('Credit Score')
axes[1, 0].set_ylabel('Degree of Membership')
axes[1, 0].legend()
axes[1, 0].grid(True)
axes[1, 0].set_ylim(-0.1, 1.1)


# Plot Loan Eligibility
axes[1, 1].plot(x_eligibility, y_eligibility_no, 'b', label='Not Eligible')
axes[1, 1].plot(x_eligibility, y_eligibility_maybe, 'g', label='Conditional')
axes[1, 1].plot(x_eligibility, y_eligibility_yes, 'r', label='Eligible')
axes[1, 1].set_title('Loan Eligibility')
axes[1, 1].set_xlabel('Eligibility Score')
axes[1, 1].set_ylabel('Degree of Membership')
axes[1, 1].legend()
axes[1, 1].grid(True)
axes[1, 1].set_ylim(-0.1, 1.1)


plt.tight_layout() # Merapikan layout subplot
plt.show()
```

Mamdani

```python
def layak_output(x):
    if 0.5 <= x <= 0.75:
        return (x - 0.5) / (0.75 - 0.5)
    elif 0.75 <= x <= 1.0:
        return (1.0 - x) / (1.0 - 0.75)
    else:
        return 0

def tidak_layak_output(x):
    if 0.0 <= x <= 0.25:
        return (x - 0.0) / (0.25 - 0.0)
    elif 0.25 <= x <= 0.5:
        return (0.5 - x) / (0.5 - 0.25)
    else:
        return 0

def fuzzy_mamdani_output(income, age, credit_score):
    inkam = {
        "low": income_low(income),
        "medium": income_med(income),
        "high": income_high(income)
    }

    umur = {
        "young": age_young(age),
        "middle": age_middle(age),
        "old": age_old(age)
    }

    kredit = {
        "low": credit_low(credit_score),
        "medium": credit_med(credit_score),
        "high": credit_high(credit_score)
    }

    rules = [
        ("low", "young", "low", "tidak_layak"),
        ("low", "young", "medium", "tidak_layak"),
        ("low", "young", "high", "layak"),
        ("low", "middle", "low", "tidak_layak"),
        ("low", "middle", "medium", "layak"),
        ("low", "middle", "high", "layak"),
        ("low", "old", "low", "tidak_layak"),
        ("low", "old", "medium", "tidak_layak"),
        ("low", "old", "high", "tidak_layak"),
        
        ("medium", "young", "low", "tidak_layak"),
        ("medium", "young", "medium", "layak"),
        ("medium", "young", "high", "layak"),
        ("medium", "middle", "low", "layak"),
        ("medium", "middle", "medium", "layak"),
        ("medium", "middle", "high", "layak"),
        ("medium", "old", "low", "tidak_layak"),
        ("medium", "old", "medium", "tidak_layak"),
        ("medium", "old", "high", "layak"),
        
        ("high", "young", "low", "layak"),
        ("high", "young", "medium", "layak"),
        ("high", "young", "high", "layak"),
        ("high", "middle", "low", "layak"),
        ("high", "middle", "medium", "layak"),
        ("high", "middle", "high", "layak"),
        ("high", "old", "low", "tidak_layak"),
        ("high", "old", "medium", "layak"),
        ("high", "old", "high", "layak"),
    ]

    x_range = np.linspace(0, 1, 100)
    output_membership = np.zeros_like(x_range)

    for income_level, age_level, credit_level, conclusion in rules:
        alpha = min(inkam[income_level], umur[age_level], kredit[credit_level])
        for i, x in enumerate(x_range):
            if conclusion == "layak":
                output_membership[i] = max(output_membership[i], min(alpha, layak_output(x)))
            else:
                output_membership[i] = max(output_membership[i], min(alpha, tidak_layak_output(x)))

    if np.sum(output_membership) == 0:
        return 0
    result = np.sum(output_membership * x_range) / np.sum(output_membership)
    return 1 if result >= 0.5 else 0
```

```python
x['ground_truth'] = x['Credit Score'].apply(lambda x: 1 if x == 'High' else 0)

x['kelayakan_mamdani'] = x.apply(lambda row: fuzzy_mamdani_output(
    row['Income'], row['Age'],
    100 if row['Credit Score'] == 'High' 
    else 50 if row['Credit Score'] == 'Medium'
    else 20
), axis=1)

x
```

```python
x[(x['ground_truth'] == 0) & (x['kelayakan_mamdani'] == 1)]
```

```python
x[(x['ground_truth'] == 1) & (x['kelayakan_mamdani'] == 0)]
```

```python
accuracy = accuracy_score(x['ground_truth'], x['kelayakan_mamdani'])
print(f'Akurasi sistem fuzzy Mamdani: {accuracy:.2%}')
```

Sugeno

```python
def fuzzy_sugeno_output(income, age, credit_score):
    inkam = {
        "low": income_low(income),
        "medium": income_med(income),
        "high": income_high(income)
    }

    umur = {
        "young": age_young(age),
        "middle": age_middle(age),
        "old": age_old(age)
    }

    kredit = {
        "low": credit_low(credit_score),
        "medium": credit_med(credit_score),
        "high": credit_high(credit_score)
    }
    
    rules = [
        ("low", "young", "low", 0.2),
        ("low", "young", "medium", 0.4),
        ("low", "young", "high", 0.6),
        ("low", "middle", "low", 0.4),
        ("low", "middle", "medium", 0.6),
        ("low", "middle", "high", 0.7),
        ("low", "old", "low", 0.1),
        ("low", "old", "medium", 0.2),
        ("low", "old", "high", 0.3),
        
        ("medium", "young", "low", 0.4),
        ("medium", "young", "medium", 0.6),
        ("medium", "young", "high", 0.8),
        ("medium", "middle", "low", 0.55),
        ("medium", "middle", "medium", 0.7),
        ("medium", "middle", "high", 0.8),
        ("medium", "old", "low", 0.3),
        ("medium", "old", "medium", 0.45),
        ("medium", "old", "high", 0.6),
        
        ("high", "young", "low", 0.6),
        ("high", "young", "medium", 0.75),
        ("high", "young", "high", 0.9),
        ("high", "middle", "low", 0.7),
        ("high", "middle", "medium", 0.8),
        ("high", "middle", "high", 0.9),
        ("high", "old", "low", 0.45),
        ("high", "old", "medium", 0.6),
        ("high", "old", "high", 0.75),
    ]

    weighted_sum = 0
    weight_total = 0

    for income_level, age_level, credit_level, output in rules:
        w = min(inkam[income_level], umur[age_level], kredit[credit_level])
        weighted_sum += w * output
        weight_total += w

    if weight_total == 0:
        return 0  

    result = weighted_sum / weight_total
    if result >= 0.5000:
        return 1
    else:
        return 0
```

```python
x['kelayakan_sugeno'] = x.apply(lambda row: fuzzy_sugeno_output(
    row['Income'], row['Age'],
    100 if row['Credit Score'] == 'High' 
    else 50 if row['Credit Score'] == 'Medium'
    else 20
), axis=1)

x
```

```python
x[(x['ground_truth'] == 0) & (x['kelayakan_sugeno'] == 1)]
```

```python
x[(x['ground_truth'] == 1) & (x['kelayakan_sugeno'] == 0)]
```

```python
accuracy = accuracy_score(x['ground_truth'], x['kelayakan_sugeno'])
print(f'Akurasi sistem fuzzy Sugeno: {accuracy:.2%}')
```
